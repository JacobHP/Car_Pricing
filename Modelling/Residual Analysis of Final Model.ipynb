```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train=pd.read_csv('training_residuals.csv', index_col=0)
test=pd.read_csv('test_residuals.csv', index_col=0)
```

```python
train['squared_res']=train['res']**2
test['squared_res']=test['res']**2
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.scatter(train['act'],train['res'])
plt.title('Training')
plt.subplot(1,2,2)
plt.scatter(test['act'],test['res'])
plt.title('Test')
```

```python
train_low=train[train['act']<100000]
test_low=test[test['act']<100000]
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.scatter(train_low['act'],train_low['res'])
plt.title('Training')
plt.subplot(1,2,2)
plt.scatter(test_low['act'],test_low['res'])
plt.title('Test')

```

```python
print('Training_rmse_low: ', round(np.sqrt(train_low['squared_res'].mean()),2))
print('Test_rmse_low: ', round(np.sqrt(test_low['squared_res'].mean()),2))

#need to advance this so that its a training_rmse_cv_low - not even convinced this is doable as cv_score requires kf which computes automatically. 
```
